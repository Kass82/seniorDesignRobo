```python
# extract the frames from the vid

import os

# create folder structure
os.makedirs("robosub_dataset/images/train", exist_ok=True)
os.makedirs("robosub_dataset/images/val", exist_ok=True)
os.makedirs("robosub_dataset/labels/train", exist_ok=True)
os.makedirs("robosub_dataset/labels/val", exist_ok=True)

print("YOLO folders ready")
```

```python
import shutil

# Move all frames to images/train
for f in os.listdir("frames"):
    if f.endswith(".jpg"):
        shutil.move(os.path.join("frames", f), "robosub_dataset/images/train/")

print("Frames moved to images/train")
```
